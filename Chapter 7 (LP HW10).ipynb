<h1 style="text-align:center; color:blue">7.11</h1>

Write the dual to the following linear program.  

```
max x + y  
2x + y ≤ 3  
x + 3y ≤ 5  
x, y ≥ 0  
```

Find the optimal solutions to both primal and dual LPs. 

-----

<font color='darkgray'>
    
<b>My Solution</b>

__Primal LP__
$$
\begin{align}
maximize: x_1 + x_2 \\
2x_1 + x_2\leq 3 \\
x_1 + 3x_2 \leq 5 \\
x_1, x_2 \geq 0
\end{align}
$$

$x_1=7/5, x_2=4/5, x_1+x_2=11/5$ 
    
    
__Dual LP__
$$
\begin{align}
minimize: 3y_1 + 5y_2 \\
2y_1 + y_2\geq 1 \\
y_1 + 3y_2 \geq 1 \\
y_1, y_2 \geq 0
\end{align}
$$

$y_1=1/5, y_2=2/5, 3y_1+5y_2=11/5$
</font>

<font color='darkorange'><b>Correct</b></font>

<h1 style="text-align:center; color:blue">7.12</h1>

For the linear program 

```
max x1 − 2x3  
x1 − x2 ≤ 1  
2x2 − x3 ≤ 1  
x1, x2, x3 ≥ 0  
```

prove that the solution (x1, x2, x3) = (3/2, 1/2, 0) is optimal. 

-----

<font color='darkgray'>
    
<b>My Solution</b>   
__Dual LP__
$$
\begin{align}
minimze: y_1 + y_2 \\
y_1 \geq 1 \\
-y_1 + 2y_2 \geq 0 \\
y_2 \geq -2 \\
y_1, y_2 \geq 0
\end{align}
$$

Dual LP's optimal solution:
$y_1 = 1, y_2 = 1/2, y_1 + y_2 = 3/2$
</font>

<font color='darkorange'><b>Correct</b></font>
