(a) NP proof
- Given a KITE problem solution, we can verify it's correctness by:
    - First building the adjacency list from the vertices V' of KITE G'=(V',E'), in which E' only contains the edges that are between vertices in V'. This step takes $O(m+n)$ time.
    - Find the end of tail of KITE, it would be the vertex with only one edge in adjacency list. There should be only one end of tail. This takes $O(m+n)$ time.
    - Find the rest of tail of KITE by going through the vertices that are connected to end of tail one by one, counting all the vertices that has only 2 edges. There should be a total of n tail vertices. This step takes $O((m+n)^2)$ time.
    - Get the rest of KITE by removing the tail. Then check if all the vertices are connected with each other. This step takes $O(n^2)$ time.
- Therefore, overall checking correctness of the solution takes $O((m+n)^2)$ time.
    
(b) Input transformation
- We can reduce Clique into a KITE problem. Given a Clique problem with input graph G=(V,E) in which we try to determine if there exists a subset of size $\geq g$ clique:

- If there is a solution to Clique problem, a solution to KITE is gaurenteed because the addition of tails to each vertex.
- If there is a solution to KITE problem, a solution to Clique problem is gaurenteed because by removing the tail.
