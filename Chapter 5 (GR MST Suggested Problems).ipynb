<h1 style="text-align:center; color:blue">5.5</h1>

Consider an undirected graph G = (V, E) with nonnegative edge weights we ≥ 0.  Suppose that you have computed a minimum spanning tree of G, and that you  have also computed shortest paths to all nodes from a particular node s ∈ V.  Now suppose each edge weight is increased by 1: the new weights are  w e = we + 1.  

(a) Does the minimum spanning tree change? Give an example where it  changes or prove it cannot change. 

(b) Do the shortest paths change? Give an example where they change or  prove they cannot change. 

-----

(a) MST does not change

- With Kruskal's algorithm, apply cut property for every edge added, for both original G and new G'.
- Since order and relation of weight of edges stay the same, a MST in G will still be MST in G'.

(b) Shortest path can change

- [A]-1-[B]-1-[C]-[D]-1-[E], [A]-5-[E]
- [A]-2-[B]-2-[C]-2-[D]-2-[E], [A]-5-[E]

-----
<font color='red'><b>OA solution Part (a)</b></font>

Let the undirected weighted graph be $G = (V, E)$ with edge weights ${w_e}_{e \in E}$, and define new weights $w’_e = w_e + 1$ for all $e \in E$.
We prove using the cut property that the minimum spanning tree (MST) of $G$ does not change.

⸻

Cut Property

For any cut $(S, V \setminus S)$, let $E(S, V \setminus S)$ be the edges crossing the cut.
The cut property states that any edge of minimum weight across a cut is safe, meaning every MST must include at least one such edge.

⸻

Proof

For any two edges $e, f \in E$:

$$
w’_e = w_e + 1, \quad w’_f = w_f + 1 \quad \Rightarrow \quad w’_e - w’_f = w_e - w_f
$$

Thus, the relative order of edge weights is preserved:

$$
w_e < w_f \iff w’_e < w’_f
$$

Hence, the minimum-weight edge(s) across every cut remain identical under $w$ and $w’$.

By the cut property, since the safe edges (minimum across cuts) are unchanged,
the set of MST edges remains exactly the same.

⸻

Conclusion

Uniformly adding a constant to all edge weights preserves the order of edges and the set of safe edges across cuts. Therefore,

$$
\text{MST}(G, w) = \text{MST}(G, w’)
$$

and the minimum spanning tree does not change.

-----
<font color='red'><b>OA solution Part (b)</b></font>

The shortest paths from $s$ can change when all edge weights are increased by $1$.

Example:
Consider a graph with vertices $s, a, b, u$ and edges:

$$
w(s,a) = 1, \quad w(a,b) = 1, \quad w(b,u) = 1, \quad w(s,u) = 4
$$
	•	Before increment:
	•	Path $s \to a \to b \to u$ has cost $1 + 1 + 1 = 3$.
	•	Direct edge $s \to u$ has cost $4$.
	•	Hence, the shortest path is $s \to a \to b \to u$.
	•	After increment ($w’_e = w_e + 1$):
	•	Path $s \to a \to b \to u$ has cost $2 + 2 + 2 = 6$.
	•	Direct edge $s \to u$ has cost $5$.
	•	Now, the shortest path is the direct edge $s \to u$.

Thus, the shortest path changes after the uniform increment.

Explanation:
A path with $k$ edges increases in cost by $k$ after the $+1$ shift per edge.
Paths with more edges are penalized more, and thus the new shortest path may favor those with fewer edges, even if their total original length was longer.

<h1 style="text-align:center; color:blue">5.6</h1>

Let G = (V, E) be an undirected graph. Prove that if all its edge weights are  distinct, then it has a unique minimum spanning tree. 

-----

- Every step in Kruskal's, apply cut property, there is only one choice of edge

-----
<font color='red'><b>OA solution</b></font>


__Step 1. Setup__

Assume, for contradiction, that there exist two distinct MSTs, $T_1$ and $T_2$.
Let $e_1 \in T_1 \setminus T_2$.
Adding $e_1$ to $T_2$ creates a unique cycle $C$ in $T_2 \cup { e_1 }$ since $T_2$ is a tree.

Because $e_1 \notin T_2$, there must exist at least one edge $e_2 \in C \cap (T_2 \setminus T_1)$;
otherwise, $e_1$ would also form a cycle in $T_1$, which is impossible.

__Step 2. Compare Edge Weights__

Since all edge weights are distinct, $w(e_1) \neq w(e_2)$.

-  Case 1: $w(e_1) < w(e_2)$
Construct
$$
T_2’ = T_2 \cup { e_1 } \setminus { e_2 }.
$$
Then
$$
w(T_2’) = w(T_2) - w(e_2) + w(e_1) < w(T_2),
$$
contradicting the minimality of $T_2$.

-  Case 2: $w(e_1) > w(e_2)$
Construct
$$
T_1’ = T_1 \cup { e_2 } \setminus { e_1 }.
$$
Then
$$
w(T_1’) = w(T_1) - w(e_1) + w(e_2) < w(T_1),
$$
contradicting the minimality of $T_1$.

In both cases, we reach a contradiction, so $T_1$ and $T_2$ cannot both be MSTs.


__Step 3. Conclusion__


__Step 4. Alternative (Cut Property Argument)__

For any cut $(S, V \setminus S)$, let $e^$ denote the lightest edge crossing the cut.
By the cut property, $e^$ must belong to every MST.

Since all edge weights are distinct, the lightest edge across any cut is unique,
so every MST must include exactly the same set of edges.
Hence, $G$ has a unique minimum spanning tree.

<h1 style="text-align:center; color:blue">5.7</h1>

Show how to find the maximum spanning tree of a graph, that is, the spanning  tree of largest total weight. 

-----

 - Since Kruskal's algorithm works for negative weight, make all edge weight w_e into -w_e

-----
<font color='red'><b>OA solution</b></font>

## (a) Algorithm

Represent the undirected weighted input graph $G=(V,E)$ in adjacency-list form with edge weights $w : E \to \mathbb{R}$.
1.  Negate the edge weights.
For every edge $e \in E$, define a new weight function
$$
w’(e) = -w(e).
$$

2.  Run Kruskal’s algorithm on $(G, w’)$.
Use the standard MST procedure:
-  Sort all edges by increasing $w’(e)$ (equivalently, by decreasing $w(e)$).
-  Iteratively add the smallest available $w’(e)$ edge to the tree that does not form a cycle.
-  Continue until $|V| - 1$ edges have been added.
Let the resulting edge set be $E_{\text{MaxST}}$.
3.  Return $E_{\text{MaxST}}$ as the maximum spanning tree of $G$ under the original weights $w$.


## (b) Justification of Correctness

For any spanning tree $T$,

$$
w’(T) = \sum_{e \in T} w’(e) = \sum_{e \in T} (-w(e)) = -\sum_{e \in T} w(e) = -w(T).
$$

Thus,

$$
\arg\min_T w’(T) = \arg\max_T w(T).
$$

Running Kruskal’s algorithm on $w’$ finds the tree minimizing $w’(T)$, which corresponds exactly to the tree maximizing $w(T)$.

By the cut property, for any cut $(S, V \setminus S)$, the lightest edge under $w’$ is the heaviest edge under $w$. Hence, the MST under $w’$ is the maximum spanning tree under $w$.


## (c) Runtime Analysis

Negating edge weights: $O(m)$
Running Kruskal’s algorithm: $O(m \log n)$

Overall runtime:

$$
O(m) + O(m \log n) = O(m \log n)
$$

Space complexity: $O(n + m)$.

<h1 style="text-align:center; color:blue">5.20</h1>

Give a linear-time algorithm that takes as input a tree and determines whether it  has a perfect matching: a set of edges that touches each node exactly once. 

-----

- Run DFS: create ancestor to decendent relationship
- Reverse graph create Gr, run DFS again for topological ordering from sink to source
- Visit each node from sink to source order
    - For u with only one edge (u,v), mark both u and v as visited
    - If v is visited, there is no perfect matching
- Repeat until all are visited

-----
<font color='red'><b>OA solution</b></font>

## (a) Algorithm

Represent the tree $T = (V, E)$ in adjacency-list form.
A perfect matching is a subset of edges $M \subseteq E$ such that each vertex is incident to exactly one edge in $M$.

1.  Parity check.

If $|V|$ is odd, return False immediately since a perfect matching requires an even number of vertices.

2.  Initialization.

Compute the degree $\deg(v)$ for each vertex and collect all leaves (vertices with $\deg(v) = 1$) into a queue $Q$.
Initialize an array $\text{matched}[v] = \text{False}$ for all $v \in V$.

3.  Leaf removal.

-  Pop a leaf $u$.
-  If $\text{matched}[u]$ is True, continue.
-  If $\text{matched}[v]$ is True, return False.
-  Add $(u, v)$ to the matching, mark both as matched, and remove them by decreasing the degree of their neighbors.
-  If any neighbor’s degree becomes $1$, enqueue it.

4.  Return.

If all vertices are matched (i.e., $|M| = |V|/2$), return True; otherwise, return False.


## (b) Justification of Correctness

Every leaf must be matched through its only incident edge, so each such edge is forced into the matching.
Once $(u, v)$ is matched, removing both vertices preserves correctness since the remaining subgraph is still a tree.
If a leaf’s neighbor is already matched, then $u$ cannot be matched, proving the tree has no perfect matching.
By induction, the algorithm correctly determines whether $T$ admits a perfect matching.


## (c) Runtime Analysis

Computing degrees and initializing leaves takes $O(n)$.
Each vertex and edge is processed at most once during leaf removal, contributing another $O(n + m)$.
Since a tree satisfies $m = n - 1$, the total runtime is:

$$
O(n + m) = O(n)
$$

The space complexity is $O(n)$.

<h1 style="text-align:center; color:blue">5.22</h1>

In this problem, we will develop a new algorithm for finding minimum spanning  trees. It is based upon the following property:  Pick any cycle in the graph, and let e be the heaviest edge in that cycle. Then  there is a minimum spanning tree that does not contain e. 

(a) Prove this property carefully.  

(b) Here is the new MST algorithm. The input is some undirected graph  G = (V, E) (in adjacency list format) with edge weights {we}.  sort the edges according to their weights  for each edge e ∈ E, in decreasing order of we:  if e is part of a cycle of G:  G = G − e (that is, remove e from G)  return G  Prove that this algorithm is correct.  

(c) On each iteration, the algorithm must check whether there is a cycle  containing a specific edge e. Give a linear-time algorithm for this task,  and justify its correctness.  

(d) What is the overall time taken by this algorithm, in terms of |E |? 

-----

(a) Consider an MST T which contains e. Removing e breaks the tree into two connected components
say S and V \S. Since all the vertices of the cycle cannot still be connected after removing e, at
least one edge, say e0 in the cycle must cross from S to V \S. However, then replacing e by e0
gives a tree T0 such that cost(T0) <= cost(T). Since T is an MST, T0 is also an MST which does
not contain e.

(b) If e is the heaviest edge in some cycle of G, then there is some MST T not containing e. However,
then T is also an MST of G−e and so we can simply search for an MST of G−e. At every step,
the algorithm creates a new graph (G − e) such that an MST of the new graph is also an MST
of the old graph (G). Hence the output of the algorithm (when the new graph becomes a tree) is
an MST of G.

(c) An undirected edge (u, v) is part of cycle i↵ u and v are in the same connected component of
G − e. Since the components can be found by DFS (or BFS), this gives a linear time algorithm.

(d) The time for sorting is O(|E| log |E|) and checking for a cycle at every step takes O(|E|) time.
Finally, we remove |E|−|V |+1 edges and hence the running time is O(|E| log |E|+(|E|−|V |)
|E|) = O(|E|2).

-----
<font color='red'><b>OA solution</b></font>

## (a) Property and Proof

Let $G=(V,E)$ be an undirected weighted graph. For any simple cycle $C \subseteq E$, let $e^\star \in C$ be the heaviest edge in that cycle.

Claim: There exists a minimum spanning tree (MST) of $G$ that does not contain $e^\star$.

__Proof (exchange argument)__

Let $T$ be any MST of $G$. If $e^\star \notin T$, we are done. Otherwise, edges of $C$ to $T$ forms exactly one cycle that includes $e^\star$.
Because $e^\star$ is the heaviest edge on $C$, there exists an edge $f \in C$, $f \neq e^\star$, with $w(f) < w(e^\star)$.
If we define

$$
T’ = T \cup {f} \setminus {e^\star},
$$

then $T’$ is also a spanning tree, and its total weight satisfies

$$
w(T’) = w(T) + w(f) - w(e^\star) < w(T).
$$

This contradicts the minimality of $T$, so no MST can contain $e^\star$. Therefore, every MST avoids the heaviest edge in any cycle.


## (b) Algorithm and Correctness

__Algorithm (Reverse-Delete)__

Input: Undirected graph $G=(V,E)$ with edge weights ${w_e}$ in adjacency-list form.
1.  Sort all edges in decreasing order of weight.
2.  For each edge $e$ in that order:
    -  If $e$ lies on a cycle of the current graph $G$, remove it (set $G = G - {e}$).
3.  Return the resulting graph $G$.

__Correctness__

When processing edges in decreasing order, any edge $e$ that lies on a cycle is guaranteed to be the heaviest edge in at least one cycle.
By the cycle property (from part a), there exists an MST that excludes such an edge, so removing $e$ preserves the existence of an MST in the remaining graph.
After all such deletions, the resulting acyclic spanning subgraph is a tree and must be an MST.

## (c) Cycle Detection

To test whether an edge $e = {u,v}$ lies on a cycle:
1.  Temporarily remove $e$ from $G$.
2.  Run BFS or DFS from $u$.
3.  If $v$ is reachable from $u$, then there exists a path between $u$ and $v$ without $e$, so $e$ is on a cycle.
4.  Otherwise, $e$ is a bridge, not part of any cycle.

This check is correct because an edge in an undirected graph lies on a cycle if and only if its endpoints remain connected when that edge is removed.
The test runs in linear time $O(|V| + |E|)$ per edge.

## (d) Runtime Analysis
-  Sorting edges in decreasing order: $O(|E| \log |E|)$.
-  For each of the $|E|$ edges, performing a BFS/DFS on the current graph costs $O(|V| + |E|)$.

Total runtime:

$$
O(|E| \log |E|) + O(|E|(|V| + |E|)) = O(|E|^2)
$$

Since $|E| \ge |V| - 1$, the overall complexity is dominated by $O(|E|^2)$.
The reverse-delete algorithm is therefore correct but less efficient than Kruskal’s or Prim’s algorithms.

<h1 style="text-align:center; color:blue">5.23</h1>

You are given a graph G = (V, E) with positive edge weights, and a minimum  spanning tree T = (V, E') with respect to these weights; you may assume G and  T are given as adjacency lists. Now suppose the weight of a particular edge  e ∈ E is modified from w(e) to a new value ˆw(e). You wish to quickly update the  minimum spanning tree T to reflect this change, without recomputing the entire  tree from scratch. There are four cases. In each case give a linear-time algorithm  for updating the tree.  

(a) e !∈ E' and ˆw(e) > w(e).  

(b) e !∈ E' and ˆw(e) < w(e).  

(c) e ∈ E' and ˆw(e) < w(e).  

(d) e ∈ E' and ˆw(e) > w(e). 

-----

(a) Since the change only increases the cost of some other spanning trees (those including e) and the
cost of T is unchanged, it is still an MST.
6
(b) We include e in the tree, thus creating a cycle. We then remove the heaviest edge e0 in the cycle,
which can be found in linear time, to get a new tree T0. It is intuitively clear that this algorithm
should work, but a rigorous proof seems surprisingly tricky. To prove this, we argue that T0
contains a least weight edge across every cut of G and is hence an MST.
Note that since the only changed edge is e, T [ {e} already includes a least weight edge across
every cut. We only removed e0 from this. However, any cut crossed by e0, must also be crossed
by at least one more edge of the cycle, which must have weight less than or equal to e0. Since this
edge is still present in T0, it contains a least weight edge across every cut.
(c) The tree is still an MST if the weight of an edge in the tree is reduced. Hence, no changes are
required.
(d) We remove e from the tree to obtain two components, and hence a cut. We then include the
lightest edge across the cut to get a new tree T0. We can now “build up” T0 using the cut
property to show that it is an MST.
Let X ✓ T0 be a set of edges that is part of some MST, and let e1 2 T0\X. Then, T0\{e1} gives
a cut which is not crossed by any edge of X and across which e1 is the lightest edge. Hence,
X [{e1} is also a part of some MST. Continuing this, we can grow X to X = T0, which must be
then an MST.

-----
<font color='orange'><b>Complete solution</b></font>

## (a) Case (a): $e \notin E’$ and $\hat w(e) > w(e)$

Algorithm. Do nothing; keep $T$.

Justification. Only spanning trees that use $e$ get (weakly) more expensive. $T$ does not use $e$, and its total weight is unchanged, so no alternative tree becomes strictly cheaper than $T$. Hence $T$ remains an MST.

Runtime. $$O(1).$$

## (b) Case (b): $e \notin E’$ and $\hat w(e) < w(e)$

Let $e=(u,v)$ and let $\hat T$ denote the updated tree.

Algorithm (cycle add–delete).

1.  Temporarily add $e$ (with weight $\hat w(e)$) to $T$, forming $T \cup {e}$, which has a unique $u$–$v$ cycle.

2.  In the tree $T$, run DFS/BFS from $u$ to get the unique path $P_{uv}$ to $v$ and record its edges’ weights; find the heaviest edge
$$
e_{\max} \in P_{uv} \quad\text{maximizing } w(\cdot).
$$

3.  If $\hat w(e) \ge w(e_{\max})$, keep $T$ (adding $e$ would not help).
Else set
$$
\hat T ;=; T \cup {e} \setminus {e_{\max}}.
$$

Justification. Adding $e$ creates exactly one cycle; by the cycle property, in any MST we may delete a heaviest edge on that cycle. If $\hat w(e) < w(e_{\max})$, replacing $e_{\max}$ by $e$ strictly decreases the total weight and yields a spanning tree, hence an MST. If not, $T$ is already optimal.

Runtime. One DFS/BFS on the tree ($|E’|=|V|-1$) and a linear scan of the path:
$$
O(|V|) ;;(\text{equivalently } O(|V|+|E’|)).
$$

## (c) Case (c): $e \in E’$ and $\hat w(e) < w(e)$

Algorithm. Do nothing; keep $T$.

Justification. Lowering an edge weight already in the MST only reduces the total weight of $T$; no other tree can become strictly cheaper than $T$ because all other trees either keep the same weight or also decrease only if they contain $e$, in which case $T$ remains at least as good. Hence $T$ is still an MST.

Runtime. $$O(1).$$

## (d) Case (d): $e \in E’$ and $\hat w(e) > w(e)$

Let $e=(u,v)$.

Algorithm (cut repair).

1.  Remove $e$ from $T$; this splits $T$ into two components $A$ and $B$. Mark the vertices of one side (say run DFS/BFS in $T\setminus{e}$ from $u$ to label $A$).

2.  Scan all edges of $G$ once and select the lightest edge crossing the cut $(A,B)$:
$$
f^\star = \arg\min {, w(f) \mid f={x,y}\in E,; x\in A,; y\in B ,}.
$$

3.  Set
$$
\hat T ;=; \big(T \setminus {e}\big) \cup {f^\star}.
$$

Justification. Removing $e$ exposes a cut $(A,B)$. By the cut property, the lightest edge crossing any cut is safe for every MST; thus adding $f^\star$ yields a minimum spanning tree consistent with the updated weights. Since only $e$’s weight changed (increased), no other cuts need be reconsidered.

Runtime. One DFS/BFS on the tree to label a component plus one linear pass over the edges:
$$
O(|V|+|E|).
$$

⸻

Summary of Runtimes
-  (a) $O(1)$
-  (b) $O(|V|)$ (DFS/BFS on the tree and max-on-path)
-  (c) $O(1)$
-  (d) $O(|V|+|E|)$ (component labeling in the tree $+$ single scan of all edges)

All are linear-time in the size of the given (adjacency-list) inputs.

<h1 style="text-align:center; color:blue">5.24</h1>

Sometimes we want light spanning trees with certain special properties. Here’s  an example.  Input: Undirected graph G = (V, E); edge weights we; subset of vertices  U ⊂ V  Output: The lightest spanning tree in which the nodes of U are leaves  (there might be other leaves in this tree as well).  (The answer isn’t necessarily a minimum spanning tree.)  Give an algorithm for this problem which runs in O(|E | log |V|) time. (Hint:  When you remove nodes U from the optimal solution, what is left?) 

-----

-----
<font color='red'><b>OA solution</b></font>

## (a) Algorithm

Let the undirected, weighted graph be $G=(V,E)$ with edge weights $w_e>0$ and a designated subset $U\subseteq V$ that must be leaves in the output spanning tree.

1.  Core subgraph.
Form the induced subgraph on the non-leaf core:
$$
V’ ;=; V\setminus U,\qquad G’ ;=; G[V’].
$$
(Feasibility check: if $G’$ is disconnected, then no spanning tree of $G$ can have every node of $U$ be a leaf; report “no solution.”)

2.  Build the core tree.
Run Kruskal’s algorithm on $G’$ (using the original weights $w$) to obtain
$$
T’ ;=; \text{MST of } G’ \text{ on vertex set } V’.
$$

3.  Attach each prescribed leaf.
For each $u\in U$, choose the lightest incident edge from $u$ to the core:
$$
e_u ;=; \arg\min{, w(u,x);:; {u,x}\in E,; x\in V’ ,}.
$$
(If $u$ has no neighbor in $V’$, there is no feasible solution.)
Add all such edges to $T’$:
$$
T ;=; T’ ;\cup; {, e_u ;:; u\in U ,}.
$$

4.  Return $T$.
This $T$ is a spanning tree of $G$ in which every $u\in U$ has degree $1$ (a leaf), and it has minimum total weight among all spanning trees with $U$ leaves.

## (b) Justification of Correctness

-  Structure of any feasible solution.
In any spanning tree $T^\star$ where all $u\in U$ are leaves, removing $U$ and their incident edges leaves a spanning tree on $V’$. Thus the core edges of any feasible solution form a spanning tree of $G’$.

-  Optimality of the core ($T’$).
Among all spanning trees on $V’$, $T’$ minimizes total weight by definition (it is the MST of $G’$). Therefore, in any optimal solution with $U$ leaves, the core must be $T’$.

-  Optimal attachment of each $u\in U$.
In any feasible solution, each $u\in U$ has exactly one incident edge, and that edge must cross the cut $\big({u},, V’\big)$. By the cut property, the lightest edge crossing this cut is safe for every MST constrained to keep $u$ as a leaf. Hence selecting, for each $u$, the lightest edge $e_u$ to $V’$ is optimal and independent across different $u$’s.

Combining these, $T=T’\cup{e_u:u\in U}$ is a minimum-weight spanning tree subject to the constraint that all nodes in $U$ are leaves.

## (c) Runtime Analysis

Let $n=|V|$, $m=|E|$.
-  Building $G’$ and (optionally) checking connectivity: one scan of adjacency lists: $$O(n+m).$$
-  Kruskal’s on $G’$ (with $m’ \le m$ edges): $$O(m’\log n)\subseteq O(m\log n).$$
-  Selecting $e_u$ for all $u\in U$: a single pass over all incident edges suffices; overall $$O(m).$$

Total:
$$
O(n+m) ;+; O(m\log n) ;+; O(m) ;=; O(m\log n).
$$

Space usage is $O(n+m)$ for the graph, sort structures, and union–find.
