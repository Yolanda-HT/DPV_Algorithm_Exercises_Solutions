__Part (d)__

We aim to show that any feasible solution $(\mathbf{x}, \mathbf{y})$ to the general dual LP must satisfy the following property: for any directed path $P$ from $s$ to $t$ in the network, the sum of the $y_{uv}$ values along the path must be at least 1.

That is, we must prove:
$$\sum_{(u, v) \in P} y_{uv} \ge 1$$

*1. Setup the Path and Dual Constraint*

Let $P$ be a directed path from $s$ to $t$, consisting of a sequence of vertices $P = (s=v_0, v_1, v_2, \dots, v_k=t)$, and a sequence of edges $e_i = (v_{i-1}, v_i)$ for $i=1, \dots, k$.

The **dual constraint** for any edge $(u, v) \in E$ is:
$$y_{uv} - z_u + z_v \ge 0$$

Rearranging this constraint gives us a lower bound on the edge cost $y_{uv}$:
$$y_{uv} \ge z_u - z_v$$

*2. Sum the Constraints Along the Path*

We sum the lower bound on $y_{uv}$ for all edges $(u, v)$ along the path $P$:

$$\sum_{(u, v) \in P} y_{uv} = \sum_{i=1}^{k} y_{v_{i-1}v_i} \ge \sum_{i=1}^{k} (z_{v_{i-1}} - z_{v_i})$$

*3. Evaluate the Telescoping Sum*

The right-hand side is a **telescoping sum** because the inner terms cancel each other out:

$$\sum_{i=1}^{k} (z_{v_{i-1}} - z_{v_i}) = (z_{v_0} - z_{v_1}) + (z_{v_1} - z_{v_2}) + \dots + (z_{v_{k-1}} - z_{v_k})$$

The intermediate terms ($z_{v_1}, z_{v_2}, \dots, z_{v_{k-1}}$) all cancel out, leaving only the terms associated with the starting vertex $v_0$ (the source $s$) and the ending vertex $v_k$ (the sink $t$):

$$\sum_{i=1}^{k} (z_{v_{i-1}} - z_{v_i}) = z_{v_0} - z_{v_k}$$

*4. Apply Boundary Conditions*
    
By definition of the general dual LP, we fix the potential variables for the source $s$ and the $\operatorname{sink} t$:
* $z_s = 1$ (The start of the path, $v_0$)
* $z_t = 0$ (The end of the path, $v_k$)

Substituting these boundary conditions:
$$z_{v_0} - z_{v_k} = z_s - z_t = 1 - 0 = 1$$

*5. Conclusion*

Combining the steps, we conclude that the sum of the $y_{uv}$ values along any directed path $P$ from $s$ to $t$ must be at least 1:

$$\sum_{(u, v) \in P} y_{uv} \ge 1$$

```
a11x1 +···+ a1nxn ≤ b1  
.  
.  
.  
am1x1 +···+ amnxn ≤ bm
```

## Shortest Path as a Minimum Cost Flow Problem

### (a) Equivalence to Minimum Cost Flow

Finding the shortest path from a source node $s$ to a terminal node $t$ is indeed equivalent to finding an $s-t$ flow $f$ that minimizes $\sum_{e} l_e f_e$ subject to $\text{size}(f) = 1$, where $l_e > 0$ are the edge lengths, and there are no capacity constraints.

* **Shortest Path to Flow:** A path $P$ from $s$ to $t$ with length $L = \sum_{e \in P} l_e$ can be seen as a flow of size 1 where $f_e = 1$ for all edges $e$ on the path and $f_e = 0$ for all other edges. The cost of this flow is $\sum_e l_e f_e = L$.
* **Flow to Shortest Path:** Because the edge lengths $l_e$ are positive and there are no capacity constraints, the flow that minimizes the total cost must be an **integral** flow (a single path), even if we initially consider fractional flows. If a minimum cost flow $f$ were fractional, it could be decomposed into a set of paths and cycles. Since $l_e > 0$, any cycle in the flow must have a positive cost, and removing this cycle would not violate the flow constraint but would reduce the total cost, which contradicts the minimality of $f$. Thus, the optimal flow $f$ of size 1 must correspond to a single $s-t$ path, and minimizing $\sum_e l_e f_e$ is equivalent to minimizing the length of that path.

---

## The Shortest Path Linear Program (Primal LP)

### (b) Writing the Shortest Path Problem as a Linear Program

We can formulate the shortest path problem from $s$ to $t$ as a minimum cost flow LP with a demand of 1 at $t$ and a supply of 1 at $s$. The variables are the flow $f_{uv}$ on each edge $(u, v) \in E$.

$$\begin{array}{rll} \text{minimize } & \sum_{(u, v) \in E} l_{uv} f_{uv} & \\ \text{subject to } & \sum_{v:(u, v) \in E} f_{uv} - \sum_{v:(v, u) \in E} f_{vu} = \begin{cases} 1 & \text{if } u = s \\ -1 & \text{if } u = t \\ 0 & \text{for all } u \in V \setminus \{s, t\} \end{cases} & (\text{Flow Conservation at node } u) \\ & f_{uv} \geq 0 & \text{for all } (u, v) \in E \end{array}$$

---

## The Dual Shortest Path Linear Program

### (c) Writing the Dual LP

The primal LP from (b) has a variable $f_{uv}$ for each edge $(u, v)$ and a constraint for each node $u$. In the dual LP, we will have a dual variable, $x_u$, corresponding to the flow conservation constraint at each node $u$.

The primal is a **minimization** problem with **equality** constraints for flow conservation and **non-negativity** constraints ($f_{uv} \geq 0$).
* The dual will be a **maximization** problem.
* The dual variables ($x_u$) are **unrestricted in sign** because the primal constraints are equalities.
* There is one dual constraint for each primal variable $f_{uv}$.

**Primal Objective:** $\sum_{(u, v) \in E} l_{uv} f_{uv}$
**Primal Constraints (for node $u$):**
$$\sum_{v:(u, v) \in E} f_{uv} - \sum_{v:(v, u) \in E} f_{vu} = b_u$$
where $b_s=1$, $b_t=-1$, and $b_u=0$ otherwise.

**Dual Objective:** $\sum_{u \in V} b_u x_u = (1)x_s + (-1)x_t + \sum_{u \in V \setminus \{s, t\}} (0)x_u = x_s - x_t$

**Dual Constraints (for primal variable $f_{uv}$):** The coefficient of $f_{uv}$ in the flow conservation constraints is:
* $+1$ in the constraint for node $u$ (source of the flow).
* $-1$ in the constraint for node $v$ (sink of the flow).
* $0$ in all other constraints.

Since the primal is **min** and the primal constraints are of the form $\geq$ (or equality), the dual constraints are of the form **(coefficients of $f_{uv}$) $\leq$ (coefficient of $f_{uv}$ in primal objective)**.

$$(1)x_u + (-1)x_v \leq l_{uv}$$
$$x_u - x_v \leq l_{uv} \quad \text{for all } (u, v) \in E$$

Thus, the dual LP is:

$$\begin{array}{rll} \text{maximize } & x_s - x_t & \\ \text{subject to } & x_u - x_v \leq l_{uv} & \text{for all } (u, v) \in E \\ & x_u \text{ unrestricted} & \text{for all } u \in V \end{array}$$

---

## Comparison with the Textbook Dual

### (d) Why the Dual LP is Different from the Textbook

The textbook (Dasgupta, *Algorithms*, page 209) is likely referring to the dual LP formulation that is used to find the **distances** from a **fixed source $s$ to all other nodes $u$** in the graph, where $x_u$ represents the length of the shortest path from $s$ to $u$, $d(s, u)$.

That textbook formulation usually looks like this:

$$\begin{array}{rll} \text{maximize } & x_t & \text{ (or } x_u \text{, if finding } d(s, u) \text{)} \\ \text{subject to } & x_v \leq x_u + l_{uv} & \text{for all } (u, v) \in E \\ & x_s = 0 & \\ & x_u \text{ unrestricted} & \end{array}$$

Your derived dual LP from (c):
$$\begin{array}{rll} \text{maximize } & x_s - x_t & \\ \text{subject to } & x_u - x_v \leq l_{uv} & \text{for all } (u, v) \in E \\ & x_u \text{ unrestricted} & \end{array}$$

**The reason for the difference is the choice of the primal LP.**

1.  **Your Primal LP (b):** Your primal is a **minimum cost flow** problem finding a flow of size 1 between *just $s$ and $t$*. The objective $\sum l_{uv} f_{uv}$ seeks the shortest $s-t$ path.
2.  **The Textbook's Dual Interpretation:** The dual on page 209 is the dual of a different primal problem: one that seeks to find a set of $\mathbf{s-u}$ paths, for all $u$, where the objective is to maximize $x_t$ (which represents the distance $d(s, t)$) subject to the **triangle inequality** constraints ($x_v \leq x_u + l_{uv}$) and a boundary condition ($x_s = 0$).

**Crucially, your dual LP and the textbook's dual are mathematically equivalent ways to express the shortest path problem.**

Let's show the connection:

* **If you set $x_s' = 0$** in your dual and let $x_u' = x_s - x_u$, then $x_u'$ represents the shortest path distance $d(s, u)$.
    * **Objective:** $x_s - x_t = x_t'$ (Maximizing the shortest path distance to $t$).
    * **Constraint:** $x_u - x_v \leq l_{uv}$ becomes $(x_s - x_v) - (x_s - x_u) \leq l_{uv}$, which simplifies to $x_u' - x_v' \leq l_{uv}$, or $x_v' \geq x_u' - l_{uv}$.

The standard textbook dual typically **sets $x_s = 0$** and then has constraints ensuring that $x_v \leq x_u + l_{uv}$, where $x_u$ is interpreted as $d(s, u)$. Your dual is more general as it doesn't fix $x_s$ to 0, which is natural when dualizing the flow formulation where the supply/demand is just an $s-t$ pair.
