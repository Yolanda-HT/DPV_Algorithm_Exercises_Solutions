<h1 style="text-align:center; color:blue">8.6</h1>

On page 251 we saw that 3SAT remains NP-complete even when restricted to  formulas in which each literal appears at most twice.  

(a) Show that if each literal appears at most once, then the problem is  solvable in polynomial time.  

(b) Show that INDEPENDENT SET remains NP-complete even in the special case  when all the nodes in the graph have degree at most 4. 

-----
<font color='darkorange'>
<b>Completed solution Part (a)</b>
    
Let the number of variables be n and the number of clauses be m. 
- Note that each variable x can satisfy at most 1 clause (because it appears as $x$ and $\neg x$ at most once). 
- We construct a bipartite graph, with the variables on the left side and the clauses on the right. 
    - Connect each variable to the clauses it appears in. 
        - For the formula to be satisfiable, each clause must pick one (at least) variable it is connected to 
        - with each variable being picked by at most one of the clauses it is connected to. 
- If we connect all the variables to a source s, all clauses to a sink t and fix the capacity of all edges as 1, this is equivalent to asking if we can send a flow of m units from s to t. 
- Since the flow problem can be solved in polynomial time, so can the satisfiability problem.
    
<b>Completed solution Part (b)</b>
    
Consider the reduction from 3SAT to INDEPENDENT SET. 
- If each literal is allowed to appear at most twice, we claim that the graph we create in this reduction has degree at most 4 for each vertex. 
- Let $(x1 \vee x2 \vee x3)$ be a clause in the original formula. 
    - Then, in the graph, the vertex corresponding to x1 in this clause, has one edge each to x2 and x3. 
    - Also, it has edges to all occurrences of $\neg x_1$. 
        - But, since $\neg x_1$ is allowed to appear at most twice, this can add at most two edges. 
    - Similarly, each vertex in the graph has at most 4 edges incident to it.
- However, we know that the special case of 3SAT, with each literal occurring at most twice is NP-complete. 
- The above argument shows that this special case reduces to the special case of INDEPENDENT SET, with each vertex having degree at most 4. Hence, this special case of INDEPENDENT SET must also be NP-complete.

- If MCS has a solution, then CLIQUE has a solution.  
- If CLIQUE has a solution, then MCS has a solution.

<h1 style="text-align:center; color:blue">8.20</h1>

In an undirected graph G = (V, E), we say D ⊆ V is a dominating set if every  v ∈ V is either in D or adjacent to at least one member of D. In the DOMINATING  SET problem, the input is a graph and a budget b, and the aim is to find a  dominating set in the graph of size at most b, if one exists. Prove that this  problem is NP-complete.

<b>Completed solution</b>
    
**(1) NP Proof**

Given an instance  
$$ (G = (V,E),\, b) $$  
and a candidate dominating set  
$$ D \subseteq V, \quad |D| \le b, $$  
we verify validity in polynomial time:

1. Check that $$|D| \le b,$$ which takes \(O(|V|)\).
2. For each vertex \(v \in V\), verify that either  
   - \(v \in D\), or  
   - \(v\) has a neighbor in \(D\).  
   We can check adjacency by scanning \(v\)'s neighbors, taking \(O(\deg(v))\).  
   Over all vertices this is \(O(|E|)\).

Thus verification takes  
$$O(|V| + |E|),$$  
which is polynomial in the input size.

Hence,  
$$ \text{DOMINATING SET} \in \mathbf{NP}. $$

**(2) Input Transformation**

We reduce **VERTEX COVER** to **DOMINATING SET**.

A VERTEX COVER instance is  
$$ (G = (V,E),\, k). $$

We construct a DOMINATING SET instance \((G',\, b)\) as follows:

- Start with \(G' = G\).  
- For every edge \((u,v) \in E\), add a new vertex \(a_{uv}\).  
- Add edges  
  $$ (u, a_{uv}), \quad (v, a_{uv}). $$
  This forms a triangle on \((u, v, a_{uv})\).

Let the resulting graph be  
$$ G' = (V', E'), \quad V' = V \cup \{a_{uv} : (u,v)\in E\}. $$

Set  
$$ b = k. $$

The construction adds one vertex and two edges per original edge, so it runs in  
$$ O(|V| + |E|). $$

Thus the input transformation is polynomial.

**(3) Output Transformation**

Suppose a DOMINATING SET solver returns a dominating set  
$$ D \subseteq V', \quad |D| \le k. $$

If the solver returns **NO**, output **NO** for VERTEX COVER.

If it returns **YES**, we convert \(D\) into a vertex cover of \(G\):

- If \(a_{uv} \in D\) for some edge \((u,v)\), replace it with either \(u\) or \(v\).  
  Since the only neighbors of \(a_{uv}\) are \(u\) and \(v\), this replacement preserves domination.

Let \(D' \subseteq V\) be the set obtained by replacing all \(a_{uv}\)'s.  
We ensure  
$$ |D'| \le |D| \le k. $$

We output  
$$ C = D' $$
as the vertex cover.

This transformation takes at most \(O(|E|)\), hence polynomial.

We prove the equivalence using the standard “both-positive’’ IFF pattern:

- **If A has a solution then B has a solution**, and  
- **If B has a solution then A has a solution**.

Here  
A = VERTEX COVER,  
B = DOMINATING SET.

---

**(⇒) If \(G\) has a vertex cover of size at most \(k\), then \(G'\) has a dominating set of size at most \(k\).**

Let \(C \subseteq V\) be a vertex cover of size at most \(k\).

We claim that \(C\) is also a dominating set of \(G'\):

- For any original vertex \(v \in V\):  
  - If \(v \in C\), it is dominated.  
  - If \(v \notin C\), since \(C\) is a vertex cover, any edge \((u,v)\) has \(u \in C\), so \(v\) is dominated by \(u\).

- For any new vertex \(a_{uv}\):  
  - Its only neighbors are \(u\) and \(v\).  
  - Since \((u,v)\) is covered, at least one of \(u\) or \(v\) is in \(C\), so \(a_{uv}\) is dominated.

Therefore \(C\) is a dominating set of \(G'\) of size at most \(k\).  
This implies  
$$ (G', b=k) \text{ is a YES instance of DOMINATING SET}. $$

---

Let  
$$ D \subseteq V', \quad |D| \le k, $$  
be a dominating set for \(G'\).

We may assume without loss of generality that \(D\) contains **no** vertices of the form \(a_{uv}\):

- If \(a_{uv} \in D\), replace it with either \(u\) or \(v\).  
- This maintains domination (since \(a_{uv}\)'s only neighbors are \(u,v\)).  
- Doing this for all such vertices yields a dominating set  
  $$ D' \subseteq V, \quad |D'| \le |D| \le k. $$

Now consider any edge \((u,v) \in E\).  
In \(G'\), we created a triangle \((u, v, a_{uv})\).  
Because \(a_{uv}\)’s only neighbors are \(u\) and \(v\), and it must be dominated by \(D'\), at least one of \(u\) or \(v\) must be in \(D'\).

Thus, for every edge of \(G\), at least one endpoint is in \(D'\).  
Hence \(D'\) is a vertex cover of size at most \(k\).

- Hence, removing the edges in F breaks this cycle.
    
<b>Completed solution Part (c)</b>
    
- We replace all the edges of the form $(w'_i, w_j)$ in the given feedback arc set, say F, by edges of the form $(w_k, w'_k)$. By the argument in the previous part, any cycle containing $(w'_i, w_j)$ must also contain $(w_i, w'_i)$ and $(w_j, w'_j)$ and hence F \ $\{ (w'_i, w_j) \cup {(w_i, w'_i)}\}$ is also a feedback arc set. Continuing in this manner, we never increase the size of the set (but might decrease it if we include the same edge, say $(w_i, w'_i)$ for two removed edges $(w'_i, w_j)$ and $(w'_i, w_k)$), we get a feedback ark set F' such that $|F' \leq b|$.
- We now claim that U = $\{v_i | (w_i, w'_i) \in F' \}$ must be a vertex cover for G. If not, then there must be an edge $(v_j, v_k)$ such that $v_j, v_k \notin U$ and hence $(w_j, w'_j), (w_k, w'_k) \notin F'$. But then $w_j \to w'_j \to w_k \to w'_k \to w_j$ would be a cycle in G' even after removing F', which is a contradiction.
